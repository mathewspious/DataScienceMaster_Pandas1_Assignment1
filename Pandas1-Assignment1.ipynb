## Assignment 1

Count distance to previous Zero

```python
import numpy as np
import pandas as pd
df = pd.DataFrame(data={'X':[7,2,0,3,4,2,5,0,3,4]})

def findDistance(x):
    position = 0
    counter = 0
    y=[]
    for i in x:
        counter = counter + 1
        if i == 0:
            position = counter
    
        y.append(counter - position)
    
    pd.Series(data=y)
    return y   

df['Y']=findDistance(df['X'])
df
```

## Assignment 4
Find highest value for 4 consicutive months

```python
import numpy as np
import pandas as pd
from datetime import datetime
from numpy.random import randn
np.random.seed(101)

dateTimeSeries=pd.date_range(start='1/1/2015', end='12/31/2015', freq='B')

h = pd.DataFrame(randn(len(dateTimeSeries)),dateTimeSeries)
h['Month']=h.index.month
h['Date']=h.index.date
h['Group'] = h['Month'].apply(lambda month : 1 if(month < 5) else(2 if(month < 9)  else 3))
h.drop('Month',axis=1,inplace=True)
h.set_index('Date',inplace=True)
h.rename(index=str, columns={0: "value"},inplace=True)
z=[]

for group in h['Group'].unique():
    maxVal =h[h.Group == group].max()['value']
    z.append(maxVal)
    
h[h['value'].isin(z)]
```
